## __Unstructred Text Analysis__


Business Question: What features/ attributes of the Hoka Clifton 9 (comfort, durability, price, etc.) are most commonly mentioned in reviews? 

This business question could help identify which attributes are most important to Hoka customers. The question also helps Hoka's marketing department figure out what customers want out of the next iteration of the shoe, along with what type of activities they're using the shoe for. Marketing can figure out how to target the needs of their customers through this analysis. 




```python
import wordcloud as wc                 
import matplotlib.pyplot as plt 
from wordcloud import STOPWORDS

Assignment5TextAnalysis = open("./PythonCSV/TextAnalysis.txt",'t+r')     # Open text file I created with customer reviews. 
txtCloudText = Assignment5TextAnalysis.read()       
Assignment5TextAnalysis.close()    

ExtraStopWordsFile = open("./PythonCSV/VoidWords.txt", 't+r')       # Open Void Words file (continuously save after editing it)
VoidWords = ExtraStopWordsFile.read()       
ExtraStopWordsFile.close()          

stop_words=set(STOPWORDS)       
stop_words.update(VoidWords.split(','))     # Set the STOPWORDS as my VoidWords text file (ensure correct formatting)

WordCloud = wc.WordCloud(width=1200,height=600,background_color='white',stopwords=stop_words).generate(txtCloudText)        # Set properties for Word Cloud and generate it
plt.figure(figsize=(12,6))     
plt.imshow(WordCloud,interpolation='bilinear')      
plt.axis("off")     
plt.show()          
```

### __Analysis Summary:__

Business Question: What features/ attributes of the Hoka Clifton 9 (comfort, durability, price, etc.) are most commonly mentioned in reviews? 

The answer to the business question is illuminated using a WordCloud. The WordCloud was generated by extracting 100 4-5 star customer reviews from Hoka's website. These reviews were helpful in showing what customers enjoyed the most about their Hoka Clifton 9's, along with what activities they use them for. At first, there was a plethora of words I considered to be "noise", so I added them to my VoidWords text file so they wouldn't appear in the WordCloud. Doing this further enhanced the signal of my WordCloud to view the important positive features of the shoe. 

First off, the most prominent word in the analysis is "comfortable". There are various other words synonomous to comfortable throughout the analysis including comfort, cushion, cushioning, support, etc. This emphasizes the importance of having a comfortable shoe for Hoka customers. With this information, Hoka can conclude that they should use the same technologies for the next iteration of the shoe and not switch anything up too drastically. Also, the marketing team can advertise the next iteration by using these key words in advertisements. 

The words run, running, everyday, and walking all appear largely on the WordCloud. This tells viewers that most customers are using this shoe for these purposes along with work and the gym. These bits of information being commonly mentioned will further help the marketing team by segmenting their target markets hobbies and/or their needs in a shoe. 

The last cluster of words that I noticed appeared prominently would be about the look of the shoe (style, stylish, color, etc.) Hoka has been known in recent years for having "trendy" shoes in their line, and the importance of style for their customers is shown in the WordCloud. Hoka's development team can use this data to improve upon crafting a stylish next iteration of the Clifton line, which consumers would appreciate.
